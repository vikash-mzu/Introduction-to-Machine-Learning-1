```python
Q1. Explain the following with an example:
A) Artificial Intelligence (AI): AI refers to the simulation of human intelligence by machines. It includes reasoning, problem-solving, language understanding, and perception.

Example: Siri and Alexa are AI systems that understand natural language and assist users with tasks.
B) Machine Learning (ML): ML is a subset of AI that enables machines to learn from data without being explicitly programmed.

Example: A recommendation system on Netflix uses ML to suggest shows based on user preferences.
C) Deep Learning (DL): DL is a subset of ML that uses neural networks with many layers to model complex patterns in data.

Example: Self-driving cars use deep learning to recognize traffic signs, pedestrians, and other vehicles.
Q2. What is supervised learning? List some examples of supervised learning.
Supervised learning is a type of machine learning where the model is trained using labeled data, meaning the input comes with the correct output. The model learns to map the input to the correct output and makes predictions for new, unseen data.

Examples:
Spam detection: Classifying emails as spam or not spam.
Image classification: Identifying objects in images (e.g., cat vs. dog).
Sentiment analysis: Predicting whether a review is positive or negative.
Q3. What is unsupervised learning? List some examples of unsupervised learning.
Unsupervised learning involves training a model on data that has no labels, meaning the algorithm tries to find patterns and relationships in the data on its own.

Examples:
Clustering: Grouping customers based on purchasing behavior.
Anomaly detection: Identifying unusual patterns in network traffic.
Dimensionality reduction: Reducing the number of features while retaining important information, such as in image compression.
Q4. What is the difference between AI, ML, DL, and Data Science (DS)?
AI (Artificial Intelligence): The broadest field, concerned with creating intelligent machines that can simulate human behavior.
ML (Machine Learning): A subset of AI focused on building algorithms that can learn from data.
DL (Deep Learning): A specialized branch of ML that uses neural networks with many layers to handle large amounts of data.
DS (Data Science): A field that encompasses data analysis, statistical modeling, and ML to extract insights from data. It uses techniques from AI and ML but is broader, often involving data cleaning, visualization, and domain knowledge.
Q5. What are the main differences between supervised, unsupervised, and semi-supervised learning?
Supervised learning: Uses labeled data, where each input has a known output, to train the model.
Example: Predicting house prices based on labeled datasets.
Unsupervised learning: Works with unlabeled data and discovers patterns or relationships within the data.
Example: Customer segmentation without predefined categories.
Semi-supervised learning: A combination of both, where the model is trained on a small amount of labeled data and a large amount of unlabeled data.
Example: A facial recognition system trained on a few labeled images and many unlabeled ones.
Q6. What is train, test, and validation split? Explain the importance of each term.
Training set: The portion of data used to train the model.
Validation set: A subset used to tune the modelâ€™s hyperparameters (e.g., learning rate).
Test set: A separate subset used to evaluate the final model's performance.
Importance:

Training set ensures the model learns the relationships in the data.
Validation set helps avoid overfitting by fine-tuning the model.
Test set provides a final check on how well the model generalizes to new data.
Q7. How can unsupervised learning be used in anomaly detection?
Unsupervised learning can identify patterns in normal data and flag any deviations as anomalies. Since anomaly detection often lacks labeled data (anomalies may be rare or undefined), unsupervised methods such as clustering or density estimation are useful in detecting unusual behavior.

Example: Detecting fraudulent transactions by identifying patterns that deviate from normal customer behavior.
Q8. List down some commonly used supervised learning algorithms and unsupervised learning algorithms.
Supervised Learning Algorithms:

Linear Regression
Decision Trees
Random Forests
Support Vector Machines (SVM)
Logistic Regression
Unsupervised Learning Algorithms:

K-means Clustering
Hierarchical Clustering
Principal Component Analysis (PCA)
Autoencoders (for deep learning tasks)
Gaussian Mixture Models (GMM)
```
